```python
import time 
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
url="https://www.amazon.com/"
driver = webdriver.Chrome(executable_path='C:\Windows\chromedriver.exe') 
driver.get(url)
driver.find_element_by_id('nav-global-location-popover-link').click()
driver.find_element_by_id("GLUXZipUpdateInput").send_keys("10081")
driver.find_elements_by_class_name("a-button-input")[3].click()
driver.find_elements_by_class_name('a-button-input')[2].click()
driver.find_element_by_id('twotabsearchtextbox').send_keys("baby milk")
driver.find_element_by_id('nav-search-submit-button').click()
```

```python
links=[]
#parse the html page
for j in range(0,6):# pass the No of pages to scrap in this for loop
    html1=driver.page_source
    soup1=BeautifulSoup(html1,'html.parser')
    anchor_tag=soup1.find_all("a",attrs={"class":"a-link-normal a-text-normal"})
    for i in range(len(anchor_tag)):
        a=anchor_tag[i]["href"]
        a="https://amazon.com/"+a
        links=links+[a]
    driver.find_elements_by_xpath("//li[@class='a-last']")[0].click()
    time.sleep(5)
```

```python
texts=[]
prices=[]
ratings=[]
rating_count=[]
asins=[]
sponsord=[]
for k in range(len(links)):
    try:
        driver.get(links[k])
        html2=driver.page_source
        soup2=BeautifulSoup(html2,'html.parser')
        # product information
        text=soup2.find("span",attrs={"class":"a-size-large product-title-word-break"})
        text=text.get_text()
        texts=texts+[text]
        # price
        try:
            price=soup2.find('span',attrs={"class":"a-size-medium a-color-price priceBlockBuyingPriceString"})
            price=price.get_text()
            prices=prices+[price]
        except:
            prices=prices+["Nan"]
        # average rating
        try:
            avg_rating=soup2.find('span',attrs={"class":"a-icon-alt"})
            avg_rating=avg_rating.get_text()
            ratings=ratings+[avg_rating]
        except:
            ratings=ratings+["Nan"]
        
        #rating count
        try:
            count=soup2.find('span',attrs={"id":"acrCustomerReviewText"})
            count=count.get_text()
            rating_count=rating_count+[count]
        except:
            rating_count=rating_count+["Nan"]
        # ASIN NUMBER
        try:
            a=driver.find_elements_by_xpath("//span")
            data=[]
            for q in range(len(a)):
                p= a[q].text
                data=data+[p]
            index=data.index('ASIN :')+1
            asin=data[index]
            asins=asins+[asin]
    #except StaleElementReferenceException:
    #    asins=asins+["Nan"]
    #except valueError:
    #    alter=soup2.find_all("td",attrs={'class':"a-size-base prodDetAttrValue"})[3].get_text() 
    #    asins=asins+[alter]
        except:
            asins=asins+["Nan"]

    # Sponsord Yes No
        try:
            driver.find_element_by_xpath("//span[@aria-label='Leave feedback on Sponsored ad']").text
            sponsord=sponsord+["Yes"]
        except:
            sponsord=sponsord+["No"]
    except:
        continue
        
        


    
```

```python
comment1 = {'text':texts,"prices":prices,"sponsord":sponsord,"ASIN":asins,"Total Rating":rating_count,"Avg rating":ratings}
df = pd.DataFrame(comment1)
```
